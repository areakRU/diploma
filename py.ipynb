```python
from turtle import shape
import scipy.io
import tensorflow as tf
from IPython.display import display, HTML
import pandas as pd
from tensorflow.python.keras.layers import Conv1D, MaxPooling1D, Dense, Dropout, Reshape, InputLayer, Flatten, GlobalMaxPooling1D 
from tensorflow.python.keras.models import Sequential
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
mat = scipy.io.loadmat(r'C:\Users\areak\Desktop\diploma\nine_movs_six_sub_cut.mat')

data_array = list()

for person_id in range(mat['nine_movs_six_sub_cut'].shape[0]):
    person_array = list()
    for movement_type in range(mat['nine_movs_six_sub_cut'][person_id][0][0][0].shape[0]):
        movement_array = list()
        for movement_id in range(mat['nine_movs_six_sub_cut'][person_id][0][0][0][movement_type][0].shape[0]):
            movement = mat['nine_movs_six_sub_cut'][person_id][0][0][0][movement_type][0][movement_id]
            movement_array.append(movement)
        person_array.append(movement_array)
    data_array.append(person_array)    

data = {'person_id':[0]*len(data_array[0][0]), 'movement_type': [0]*len(data_array[0][0]), 'data': data_array[0][0]}
df = pd.DataFrame(data)
for person_id in range(len(data_array)):
    for movement_type in range(len(data_array[person_id])):
        data = {'person_id':[person_id]*len(data_array[person_id][movement_type]), 'movement_type': [movement_type]*len(data_array[person_id][movement_type]), 'data': data_array[person_id][movement_type]}
        df_temp = pd.DataFrame(data)
        df = pd.concat([df,df_temp], ignore_index=True)
df.drop(labels=range(0,52), inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
X = np.asarray(df.data.tolist())
y = pd.get_dummies(np.asarray(df.movement_type))
```

```python
print('X.shape: {0}'.format(X.shape))
print('y.shape: {0}'.format(y.shape))
display(X)
display(y)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print('X_train.shape: {0}'.format(X_train.shape))
print('X_val.shape: {0}'.format(X_val.shape))
print('Y_train.shape: {0}'.format(Y_train.shape))
print('Y_val.shape: {0}'.format(Y_val.shape))
display(X_train)
display(X_val)
display(Y_train)
display(Y_val)
```

```python
classes = 9

model = Sequential()
model.add(InputLayer(input_shape=(400,)))
model.add(Reshape((400,1), input_shape=(400,)))
model.add(Conv1D(100,30,activation='relu',input_shape=(400,1)))
model.add(Conv1D(100,30,activation='relu'))
model.add(Conv1D(100,30,activation='relu'))
model.add(MaxPooling1D(3))
model.add(Conv1D(200,30,activation='relu'))
model.add(Conv1D(200,30,activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(300, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(classes, activation='softmax'))

model.compile(optimizer = 'Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
EPOCHS = 30
history = model.fit(X_train, Y_train, epochs=EPOCHS, verbose=2,validation_data=(X_val, Y_val))
plt.figure(figsize=(10,6))
plt.plot(history.history['accuracy'], label = 'train_accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
print(history.history['val_accuracy'])
```
